```python
def conver_to_list_StandardScaler(row):
    """
    原来格式为：str  '[9, 7, 2, 1, 3, 5, 10, 16, 16, 33, 37, 31, 42, 37, 42, 37, 27, 23, 23, 24, 24, 26, 24, 24]'
    将其转为 list 且内容是标准化后的
    """
    arr = row.replace('[','').replace(']','').replace(' ','').split(',')
    arr = list(map(float,arr))
    # 这里需要将其转为array数组，因为StandardScaler().fit_transform(X=arr)需要列数据
    arr = np.array(arr).reshape(-1, 1)
    ys_standard = StandardScaler().fit_transform(X=arr)
    ys_standard_list = []
    for i in ys_standard:
        ys_standard_list.append(i[0])
    
    return ys_standard_list
```

```python
def conver_to_list_mean(row):
    """
    原来格式为：str  '[9, 7, 2, 1, 3, 5, 10, 16, 16, 33, 37, 31, 42, 37, 42, 37, 27, 23, 23, 24, 24, 26, 24, 24]'
    将其转为 list 且值是求均值的结果
    """
    #row = df['each_hour_driving_time'][55]
    arr = row.replace('[','').replace(']','').replace(' ','').split(',')
    arr = list(map(float,arr))
    sumarr = sum(arr)

    ys_standard_list = []
    for i in arr:
        if i !=0:
            ys_standard_list.append(i/sumarr)
        else:
            ys_standard_list.append(0)
    return ys_standard_list
```
