```python
def get_keyword_prob(df,keywords,which_day=0):
    '''
     返回概率  例如出现“小学” 的概率为0.12  及共100天 有12天出现在小学的地址
    :param keywords  -> list:  关键字列表
    :param which_day:  0：全部 1：周中 2：周末
    :return:
    '''
    df = df.dropna(subset=['start_address_name', 'end_address_name', 'user_id', 'car_id', 'datetime'])

    if which_day == 1:
        df = df[df['weekday'].isin(range(1, 6))]
    elif which_day == 2 :
        df = df[df['weekday'].isin([6,7])]
    else :
        df = df

    driving_workday_sum = len(set(df['datetime']))
    
    count_day = 0
    last_day = ''
    for keyword in keywords:
        if keyword in df['start_address_name'].iloc[0] or keyword in df['end_address_name'].iloc[0]:
            print('第一个')
            print(df['start_address_name'].iloc[0],df['end_address_name'].iloc[0])
            count_day += 1
            last_day = df['datetime'].iloc[0]
            break
    for i in range(1,df.shape[0]):
        #当天统计过 则不统计
        for keyword in keywords:
            if df['datetime'].iloc[i] == last_day:
                break
            if keyword in df['start_address_name'].iloc[i] or keyword in df['end_address_name'].iloc[i]:
                count_day += 1
                last_day = df['datetime'].iloc[i]
                print(df['start_address_name'].iloc[i],df['end_address_name'].iloc[i])
                break
    print(df['user_id'].iloc[0])
    return round(count_day/driving_workday_sum,2)
```
